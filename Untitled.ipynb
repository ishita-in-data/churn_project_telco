```python
# churn analysis with monthly charge
churn.groupby(pd.cut(churn['MonthlyCharges'], bins=5))['Churn'].value_counts(normalize=True)
```

```python
# there is a slight drop at last bin
# sample size effect maybe?
churn['MonthlyCharges_bin'] = pd.cut(churn['MonthlyCharges'], bins=5)
churn['MonthlyCharges_bin'].value_counts().sort_index()
```

```python
# For high-charge customers, what services do they have?
high_charge = churn[churn['MonthlyCharges'] > 90]
high_charge[['InternetService', 'OnlineSecurity', 'TechSupport', 'StreamingTV', 'MonthlyCharges', 'Churn']].head(10)
```
