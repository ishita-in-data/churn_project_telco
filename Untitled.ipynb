```python
churn['TotalCharges']=pd.to_numeric(churn['TotalCharges'], errors='coerce')
```

```python
# drops 11 rows with missing values in TotalCharges column
churn_clean=churn.dropna()
```

```python
churn_clean.info()
```

```python
churn_clean.drop(['MonthlyCharges_bin','customerID'], axis=1, inplace=True)
```

```python
num_cols=churn_clean.select_dtypes(include=['int64', 'float64']).columns
churn_clean['Churn_binary'] = (churn_clean['Churn'] == 'Yes').astype(int)
```

```python
correlation_matrix=churn_clean[list(num_cols) + ['Churn_binary']].corr()
print(correlation_matrix)
```

```python
correlation_matrix_sorted=correlation_matrix.sort_values('Churn_binary')
print(correlation_matrix_sorted)
```

```python
add_on = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
churn_clean['ServiceCount'] = (churn_clean[add_on] == 'Yes').sum(axis=1)
print(churn_clean['ServiceCount'].value_counts().sort_index())
```

```python
service_count=churn_clean.groupby('ServiceCount')['Churn'].value_counts(normalize=True).unstack()
print(service_count)
```

```python
for service in add_on:
    print(churn_clean.groupby(service)['Churn'].value_counts(normalize=True))
```

```python
demographics=['SeniorCitizen', 'Partner', 'Dependents', 'gender']
for demo in demographics:
    print(churn_clean.groupby(demo)['Churn'].value_counts(normalize=True))
```

```python
churn_clean.groupby('PaperlessBilling')['Churn'].value_counts(normalize=True)
```

```python
fig=px.pie(churn_clean, names='Churn',
             title='Churn Distribution', 
             color_discrete_sequence=px.colors.sequential.RdBu, 
             hole=0.3)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
tenure_churn_pct = (tenure_churn['Yes'] * 100).reset_index()
tenure_churn_pct.columns = ['tenure', 'churn_rate']
fig=px.line(tenure_churn_pct, x='tenure', y='churn_rate',
             title='Churn Rate Decline Over Tenure', 
             labels={'x': 'Tenure (months)', 'y': 'Churn Rate', 'color': 'Churn'},
             color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
contract_churn_pct = (contract_churn['Yes'] * 100).reset_index()
contract_churn_pct.columns = ['Contract', 'churn_rate']
contract_churn_pct = contract_churn_pct.sort_values('churn_rate', ascending=False)
fig=px.bar(contract_churn_pct, x='Contract', y='churn_rate',
             title='Churn Rate by Contract Type',
             labels={'x': 'Contract Type', 'y': 'Churn Rate', 'color': 'Churn'},
             color='Contract',
             color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
servicecount_churn_pct = (service_count['Yes'] * 100).reset_index()
servicecount_churn_pct.columns = ['ServiceCount', 'churn_rate']
fig= px.line(servicecount_churn_pct, x='ServiceCount', y='churn_rate',
             title='Churn Rate by Number of Services',
             labels={'x': 'Number of Services', 'y': 'Churn Rate', 'color': 'Churn'},
             color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_traces(fill='tozeroy', line_shape='spline')
fig.show()

```

```python
fig=px.imshow(correlation_matrix_sorted, 
             title='Correlation Matrix of Numerical Features and Churn',
             labels=dict(x="Features", y="Features", color="Correlation"),
             color_continuous_scale=px.colors.diverging.RdBu,
             color_continuous_midpoint=0,
             text_auto='.3f')
fig.show()
```

```python
feature_gaps = {}
features_to_check = ['tenure', 'Contract', 'ServiceCount', 'InternetService', 
                     'PaymentMethod', 'MonthlyCharges', 'OnlineSecurity', 
                     'TechSupport', 'SeniorCitizen', 'Dependents', 'Partner',
                     'OnlineBackup', 'DeviceProtection', 'PaperlessBilling']

for feature in features_to_check:
        feature_churn = churn_clean.groupby(feature)['Churn_binary'].apply(lambda x: (x.sum() / len(x) * 100))
        gap = feature_churn.max() - feature_churn.min()
        feature_gaps[feature] = gap
gap_df = pd.DataFrame(list(feature_gaps.items()), columns=['Feature', 'Churn_Gap_pp'])

gap_df = gap_df.sort_values('Churn_Gap_pp', ascending=True).tail(10)
print(gap_df)
```
